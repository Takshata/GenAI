```python
from easyocr import Reader
```

```python
load_hi_en_mr=Reader(['hi','en','mr'])
```

```python
load_es=Reader(['es'])
```

```python
load_fr=Reader(['fr'])
```

```python
def read_text(img_name,model_name,in_line=True):

  text=model_name.readtext(img_name,detail=0,paragraph=in_line)

  return '\n'.join(text)
```

```python
print(read_text('/content/billboard.png',load_hi_en_mr))
```

```python
import easyocr
Reader=easyocr.Reader(['en'])
result=Reader.readtext('/content/billboard2.png')

for (bbox,text,prob) in result:
  print(f'Text:{text} - probability {prob}')
```
